```python
import numpy as np

# Apply log transformation method
c = 255 / np.log(1 + np.max(simualation_img))
log_image = c * (np.log(simualation_img + 1))
   
# Specify the data type so that
# float value will be converted to int
log_image = np.array(log_image, dtype = np.uint8)
   
# Display both images
plt.imshow(simualation_img)
plt.show()
plt.imshow(log_image)
plt.show()
```

```python
#power transformation

# Apply Gamma=2.2 on the normalised image and then multiply by scaling constant (For 8 bit, c=255)
gamma_two_point_two = np.array(255*(simualation_img/255)**2.2,  dtype='uint8')

# Similarly, Apply Gamma=0.4 
gamma_point_four = np.array(255*(simualation_img/255)**0.4,dtype='uint8')
plt.imshow(gamma_two_point_two)
plt.show()
plt.imshow(gamma_point_four)
plt.show()
```

```python
# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1)*pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2
  
# Define parameters.
r1 = 70
s1 = 0
r2 = 150
s2 = 255
  
# Vectorize the function to apply it to each value in the Numpy array.
pixelVal_vec = np.vectorize(pixelVal)
  
# Apply contrast stretching.
contrast_stretched = pixelVal_vec(simualation_img, r1, s1, r2, s2)
plt.imshow(contrast_stretched)
```

```python
hist,bins = np.histogram(simualation_img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(simualation_img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
```
